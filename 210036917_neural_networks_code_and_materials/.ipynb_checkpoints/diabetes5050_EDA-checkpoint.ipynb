```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing dataset
diabetes_df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
```

```python
diabetes_df.head(10)
```

Due to the extremely low number of prediabetes-classed instances, the prediabetes and diabetes classes will be combined into one class. The recall for both of these classes are going to be one of the main metrics focussed on when training and optimising the models, which is why these two particular classes have been selected.

```python
diabetes_df.loc[diabetes_df['Diabetes_012'] == 2., 'Diabetes_012'] = 1. # combining classes
diabetes_df.rename(columns={'Diabetes_012':'Diabetes'}, inplace=True) # renaming column
```

```python
# Checking for, and removing, attributes that only show 1 value
for col_name in diabetes_df.columns:
    if diabetes_df[col_name].nunique()==1:
        diabetes_df.drop(columns=[col_name], inplace=True)
        print(f'{col_name} attribute dropped from dataset')
```

```python
y_class_instances = []
y_class_instances.append(((diabetes_df['Diabetes']==0).sum()))
y_class_instances.append(((diabetes_df['Diabetes']==1).sum()))
x_vals = ['No Diabetes', 'Prediabetes or Diabetes']
x_axis = np.arange(len(x_vals))

plt.figure(figsize=(12, 9))
plt.bar(x_vals, y_class_instances)
plt.xlabel('Target Classes')
plt.ylabel('Number of Instances')
plt.title('Target Class breakdown')
plt.savefig(f'EDA_graphs/target_class_bar_breakdown.svg', bbox_inches='tight')
plt.show()
```

```python
print(f'{y_class_instances[0]} instances of no diabetes.')
print(f' {y_class_instances[1]} instances of prediabetes or diabetes.')
```

```python
# Checking for missing values in rows
diabetes_df[diabetes_df.isnull().any(axis=1)]
```

```python
def discrete_attribute_breakdown(df, col_name):
    x_vals = np.sort(df[col_name].unique()).astype(int)
    y_no_diabetes = []
    y_diabetes = []
    for x in x_vals:
        y_no_diabetes.append(((df['Diabetes']==0) & (df[col_name]==x)).sum())
        y_diabetes.append(((df['Diabetes']==1) & (df[col_name]==x)).sum())
    x_axis = np.arange(len(x_vals))
    
    plt.figure(figsize=(12, 9))
    plt.bar(x_axis - 0.2, y_no_diabetes, 0.4, label = 'No diabetes', color='blue')
    plt.bar(x_axis + 0.2, y_diabetes, 0.4, label = 'Prediabetes or diabetes', color='red')
    if len(x_vals)>30:
        plt.xticks(x_axis[::4], x_vals[::4]) # if the number of unique values is large, reduce the number of displayed x-ticks
    else:
        plt.xticks(x_axis, x_vals)
    plt.xlabel(col_name)
    plt.ylabel('Number of Instances')
    plt.title(f'{col_name} Breakdown by Target Class')
    plt.legend()
    plt.savefig(f'EDA_graphs/{col_name}_bar_breakdown.svg', bbox_inches='tight')
    plt.show()
```

```python
# Plotting bar graph breakdown for each attribute
for col_name in diabetes_df.columns[1:]:
    discrete_attribute_breakdown(diabetes_df, col_name)
```
