```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing dataset
fetal_health_df = pd.read_csv('dataset-of-80s.csv')
```

```python
fetal_health_df
```

```python
# Checking for, and removing, attributes that only show 1 value
for col_name in fetal_health_df.columns:
    if fetal_health_df[col_name].nunique()==1:
        fetal_health_df.drop(columns=[col_name], inplace=True)
```

```python
y_class_instances = []
y_class_instances.append(((fetal_health_df['target']==0).sum()))
y_class_instances.append(((fetal_health_df['target']==1).sum()))
x_vals = ['No Hit', 'Hit']
x_axis = np.arange(len(x_vals))

plt.figure(figsize=(12, 9))
plt.bar(x_vals, y_class_instances)
plt.xlabel(col_name)
plt.ylabel('Number of Instances')
plt.title('Target Class breakdown')
#plt.savefig(f'EDA_graphs/target_class_bar_breakdown.svg', bbox_inches='tight')
plt.show()
```

```python
# Checking for missing values in rows
fetal_health_df[fetal_health_df.isnull().any(axis=1)]
```

```python
def attribute_breakdown(df, col_name):
    df.groupby('target')[col_name].plot(kind='kde')
    plt.legend()
    plt.title(f'{col_name} Density')
    plt.xlabel(f'{col_name}')
    #plt.savefig(f'EDA_graphs/{col_name}_kde.jpg', bbox_inches='tight')
    plt.show()
```

```python
for col_name in fetal_health_df.columns[3:-1]:
    attribute_breakdown(fetal_health_df, col_name)
```

```python
def numerically_encode_col(df, col_name):
    unique_vals = df[col_name].unique() # finding all unique values in the col
    num_unique = df[col_name].nunique() # finding number of unique values
    new_vals = [*range(num_unique)] # acquiring number values to map attribute classes to
    mapping_dict = dict(zip(unique_vals, new_vals)) # creating mapping dictionary
    print(col_name, 'mapping:', mapping_dict) # printing mapping to be noted down
    df[col_name] = df[col_name].map(mapping_dict) # mapping column values
    return df
```
