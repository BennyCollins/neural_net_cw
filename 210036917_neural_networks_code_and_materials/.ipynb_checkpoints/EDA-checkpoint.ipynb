```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing dataset
body_performance_df = pd.read_csv('bodyPerformance.csv')
```

```python
body_performance_df.head(10)
```

```python
# Checking for, and removing, attributes that only show 1 value
for col_name in body_performance_df.columns:
    if body_performance_df[col_name].nunique()==1:
        body_performance_df.drop(columns=[col_name], inplace=True)
```

```python
y_class_instances = []
y_class_instances.append(((body_performance_df['class']=='A').sum()))
y_class_instances.append(((body_performance_df['class']=='B').sum()))
y_class_instances.append(((body_performance_df['class']=='C').sum()))
y_class_instances.append(((body_performance_df['class']=='D').sum()))
x_vals = ['A', 'B', 'C', 'D']
x_axis = np.arange(len(x_vals))

plt.figure(figsize=(12, 9))
plt.bar(x_vals, y_class_instances)
plt.xlabel(col_name)
plt.ylabel('Number of Instances')
plt.title('Target Class breakdown')
#plt.savefig(f'EDA_graphs/target_class_bar_breakdown.svg', bbox_inches='tight')
plt.show()
```

```python
x_vals = ['A', 'B', 'C', 'D']
y_male = []
y_female = []
for x in x_vals:
    y_male.append(((body_performance_df['class']==x) & (body_performance_df['gender']=='M')).sum())
    y_female.append(((body_performance_df['class']==x) & (body_performance_df['gender']=='F')).sum())
x_axis = np.arange(len(x_vals))
plt.figure(figsize=(12, 9))
plt.bar(x_axis - 0.2, y_male, 0.4, label = 'Male')
plt.bar(x_axis + 0.2, y_female, 0.4, label = 'Female')
plt.xticks(x_axis, x_vals)
plt.xlabel(col_name)
plt.ylabel('Number of Instances')
plt.title('Class Breakdown by Gender')
plt.legend()
#plt.savefig(f'EDA_graphs/{col_name}_bar_breakdown.svg', bbox_inches='tight')
plt.show()
```

```python
# Checking for missing values in rows
body_performance_df[body_performance_df.isnull().any(axis=1)]
```

```python
def attribute_breakdown(df, col_name):
    df.groupby('class')[col_name].plot(kind='kde')
    plt.legend()
    plt.title(f'{col_name} Density')
    plt.xlabel(f'{col_name}')
    #plt.savefig(f'EDA_graphs/{col_name}_kde.jpg', bbox_inches='tight')
    plt.show()
```

```python
for col_name in body_performance_df.columns[2:-1]:
    attribute_breakdown(body_performance_df, col_name)
```

```python
def numerically_encode_col(df, col_name):
    unique_vals = df[col_name].unique() # finding all unique values in the col
    num_unique = df[col_name].nunique() # finding number of unique values
    new_vals = [*range(num_unique)] # acquiring number values to map attribute classes to
    mapping_dict = dict(zip(unique_vals, new_vals)) # creating mapping dictionary
    print(col_name, 'mapping:', mapping_dict) # printing mapping to be noted down
    df[col_name] = df[col_name].map(mapping_dict) # mapping column values
    return df
```

```python
for col_name in ['gender', 'class']:
    body_performance_df = numerically_encode_col(body_performance_df, col_name)
```

```python
body_performance_df.to_csv('processed_body_performance.csv', index=False)
```
