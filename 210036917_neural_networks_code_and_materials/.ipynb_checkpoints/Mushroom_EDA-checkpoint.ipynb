```python
# mapping letters to values for noting value-to-number encoding at the end of EDA
mushroom_df['class'] = mushroom_df['class'].map({'e':'edible','p':'poisonous'})
mushroom_df['cap-shape'] = mushroom_df['cap-shape'].map({'b':'bell','c':'conical','x':'convex','f':'flat','k':'knobbed','s':'sunken'})
mushroom_df['cap-surface'] = mushroom_df['cap-surface'].map({'f':'fibrous','g':'grooves','y':'scaly','s':'smooth'})
mushroom_df['cap-color'] = mushroom_df['cap-color'].map({'n':'brown','b':'buff','c':'cinnamon','g':'gray','r':'green','p':'pink','u':'purple','e':'red','w':'white','y':'yellow'})
mushroom_df['bruises'] = mushroom_df['bruises'].map({'t':'bruises','f':'no'})
mushroom_df['odor'] = mushroom_df['odor'].map({'a':'almond','l':'anise','c':'creosote','y':'fishy','f':'foul','m':'musty','n':'none','p':'pungent','s':'spicy'})
mushroom_df['gill-attachment'] = mushroom_df['gill-attachment'].map({'a':'attached','d':'descending','f':'free','n':'notched'})
mushroom_df['gill-spacing'] = mushroom_df['gill-spacing'].map({'c':'close','w':'crowded','d':'distant'})
mushroom_df['gill-size'] = mushroom_df['gill-size'].map({'b':'broad','n':'narrow'})
mushroom_df['gill-color'] = mushroom_df['gill-color'].map({'k':'black','n':'brown','b':'buff','h':'chocolate','g':'gray','r':'green','o':'orange','p':'pink','u':'purple','e':'red','w':'white','y':'yellow'})
mushroom_df['stalk-shape'] = mushroom_df['stalk-shape'].map({'e':'enlarging','t':'tapering'})
mushroom_df['stalk-root'] = mushroom_df['stalk-root'].map({'b':'bulbous','c':'club','u':'cup','e':'equal','z':'rhizomorphs','r':'rooted','?':'unknown'})
mushroom_df['stalk-surface-above-ring'] = mushroom_df['stalk-surface-above-ring'].map({'f':'fibrous','y':'scaly','k':'silky','s':'smooth'})
mushroom_df['stalk-surface-below-ring'] = mushroom_df['stalk-surface-below-ring'].map({'f':'fibrous','y':'scaly','k':'silky','s':'smooth'})
mushroom_df['stalk-color-above-ring'] = mushroom_df['stalk-color-above-ring'].map({'n':'brown','b':'buff','c':'cinnamon','g':'gray','o':'orange','p':'pink','e':'red','w':'white','y':'yellow'})
mushroom_df['stalk-color-below-ring'] = mushroom_df['stalk-color-below-ring'].map({'n':'brown','b':'buff','c':'cinnamon','g':'gray','o':'orange','p':'pink','e':'red','w':'white','y':'yellow'})
mushroom_df['veil-type'] = mushroom_df['veil-type'].map({'p':'partial','u':'universal'})
mushroom_df['veil-color'] = mushroom_df['veil-color'].map({'n':'brown','o':'orange','w':'white','y':'yellow'})
mushroom_df['ring-number'] = mushroom_df['ring-number'].map({'n':'none','o':'one','t':'two'})
mushroom_df['ring-type'] = mushroom_df['ring-type'].map({'c':'cobwebby','e':'evanescent','f':'flaring','l':'large','n':'none','p':'pendant','s':'sheathing','z':'zone'})
mushroom_df['spore-print-color'] = mushroom_df['spore-print-color'].map({'k':'black','n':'brown','b':'buff','h':'chocolate','r':'green','o':'orange','u':'purple','w':'white','y':'yellow'})
mushroom_df['population'] = mushroom_df['population'].map({'a':'abundant','c':'clustered','n':'numerous','s':'scattered','v':'several','y':'solitary'})
mushroom_df['habitat'] = mushroom_df['habitat'].map({'g':'grasses','l':'leaves','m':'meadows','p':'paths','u':'urban','w':'waste','d':'woods'})
```

```python
# Checking for, and removing, attributes that only show 1 value
for col_name in mushroom_df.columns:
    if mushroom_df[col_name].nunique()==1:
        mushroom_df.drop(columns=[col_name], inplace=True)
```

```python
y_class_instances = []
y_class_instances.append(((mushroom_df['class']=='poisonous').sum()))
y_class_instances.append(((mushroom_df['class']=='edible').sum()))
x_vals = ['Poisonous', 'Edible']
x_axis = np.arange(len(x_vals))

plt.figure(figsize=(12, 9))
plt.bar(x_vals, y_class_instances)
#plt.xticks(x_axis, x_vals)
plt.xlabel(col_name)
plt.ylabel('Number of Instances')
plt.title('Target Class breakdown')
plt.savefig(f'EDA_graphs/target_class_bar_breakdown.svg', bbox_inches='tight')
plt.show()
```

```python
def attribute_breakdown(col_name):
    x_vals = mushroom_df[col_name].unique()
    y_poisonous = []
    y_edible = []
    for x in x_vals:
        y_poisonous.append(((mushroom_df['class']=='poisonous') & (mushroom_df[col_name]==x)).sum())
        y_edible.append(((mushroom_df['class']=='edible') & (mushroom_df[col_name]==x)).sum())
    x_axis = np.arange(len(x_vals))
    
    plt.figure(figsize=(12, 9))
    plt.bar(x_axis - 0.2, y_poisonous, 0.4, label = 'Poisonous', color='red')
    plt.bar(x_axis + 0.2, y_edible, 0.4, label = 'Edible', color='green')
    plt.xticks(x_axis, x_vals)
    plt.xlabel(col_name)
    plt.ylabel('Number of Instances')
    plt.title(f'{col_name} breakdown')
    plt.legend()
    plt.savefig(f'EDA_graphs/{col_name}_bar_breakdown.svg', bbox_inches='tight')
    plt.show()
```

```python
for col_name in mushroom_df.columns[1:]:
    attribute_breakdown(col_name)
```

Neural networks require numerical inputs, so the class values are mapped to numerical class values.

```python
def numerically_encode_col(df, col_name):
    unique_vals = df[col_name].unique() # finding all unique values in the col
    num_unique = df[col_name].nunique() # finding number of unique values
    new_vals = [*range(num_unique)] # acquiring number values to map attribute classes to
    mapping_dict = dict(zip(unique_vals, new_vals)) # creating mapping dictionary
    print(col_name, 'mapping:', mapping_dict) # printing mapping to be noted down
    df[col_name] = df[col_name].map(mapping_dict) # mapping column values
    return df
```

Each mapping below will be noted so that contextual analysis will be possible when looking at the results of the algorithms.

```python
for col_name in mushroom_df.columns:
    mushroom_df = numerically_encode_col(mushroom_df, col_name)
```
